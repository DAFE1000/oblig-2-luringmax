```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from sympy import symbols, exp, atan, diff, lambdify

# 1. Symbolsk derivasjon
x = symbols('x')
f_expr = exp(-x/4) * atan(x)
f_prime_expr = diff(f_expr, x)
print("Derivert uttrykk:")
print(f_prime_expr)

# 2. Numerisk løsning av likningen: atan(x) - 4 / (x^2 + 1) = 0
def equation(x):
    return np.arctan(x) - 4 / (x**2 + 1)

x_kritisk = fsolve(equation, x0=1.0)[0]

# 3. Evaluer funksjonen i toppunktet
f_func = lambdify(x, f_expr, 'numpy')
f_max = f_func(x_kritisk)

print("\nKritisk punkt (x > 0):")
print(f"x = {x_kritisk:.10f}")
print("\nf(x) i dette punktet:")
print(f"f(x) = {f_max:.10f}")

# 4. Plotting
x_vals = np.linspace(-2, 10, 500)
y_vals = f_func(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = $e^{-x/4}\\cdot \\arctan(x)$', color='blue')
plt.plot(x_kritisk, f_max, 'ro', label=f'Maksimumspunkt ({x_kritisk:.4f}, {f_max:.4f})')
plt.title('Graf av funksjonen $f(x) = e^{-x/4}\\cdot \\arctan(x)$')
plt.xlabel('x-verdi')
plt.ylabel('f(x)-verdi')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
